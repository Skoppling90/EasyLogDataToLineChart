```python
dato='24-10-2018'
tidStart='14:10:00'
tidSlutt='15:46:00'
files=2
NameFile1 = '24okt_EasyLog1_Vindu.xls'
NameFile2 = '24okt_EasyLog2_Over_dør.xls'
NameFile3 = '24okt_EasyLog3_Bak_maskin.xls'
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas import ExcelWriter
from pandas import ExcelFile

if files == 1:
    df1=pd.read_excel(NameFile1)
elif files == 2:
    df1=pd.read_excel(NameFile1)
    df2=pd.read_excel(NameFile2)
elif files == 3:
    df1=pd.read_excel(NameFile1)
    df2=pd.read_excel(NameFile2)   
    df3=pd.read_excel(NameFile3)
else:
    print('files must be integer between 1 and 3')
```

```python
if files == 1:
    print(df1.head())
elif files == 2:
    print(df1.head())
    print(df2.head())
elif files == 3:
    print(df1.head())
    print(df2.head())
    print(df3.head())
```

```python
DatoTidStart = dato + ' ' + tidStart
DatoTidSlutt = dato + ' ' + tidSlutt
```

```python
if files == 1:
    df1 = df1.iloc[:,0:5]
    print(df1)
elif files == 2:
    df1=df1.iloc[:,0:5]
    df2=df2.iloc[:,0:5]
    print(df1)
    print(df2)
elif files == 3:
    df1=df1.iloc[:,0:5]
    df2=df2.iloc[:,0:5]
    df3=df3.iloc[:,0:5]
    print(df1)
    print(df2)
    print(df3)

#df1=df1.iloc[:,0:5] # fjerner alle kolonner utenom dato, tid og H2O2
```

```python
if files == 1:
    #df1=df1.drop(['EasyLog USB1'], axis=1)
    df1=df1.drop([df1.T.index[0]], axis=1)
    df1=df1.drop(['Dew Point(°C)'], axis=1)
elif files == 2:
    df1=df1.drop([df1.T.index[0]], axis=1)
    df1=df1.drop(['Dew Point(°C)'], axis=1)
    df2=df2.drop([df2.T.index[0]], axis=1)
    df2=df2.drop(['Dew Point(°C)'], axis=1)
elif files == 3:
    df1=df1.drop([df1.T.index[0]], axis=1)
    df1=df1.drop(['Dew Point(°C)'], axis=1)
    df2=df2.drop([df2.T.index[0]], axis=1)
    df2=df2.drop(['Dew Point(°C)'], axis=1)
    df3=df3.drop([df3.T.index[0]], axis=1)
    df3=df3.drop(['Dew Point(°C)'], axis=1)
```

```python
if files == 1:
    df1=df1[(df1['Time'] > DatoTidStart) & (df1['Time'] < DatoTidSlutt)]
elif files == 2:
    df1=df1[(df1['Time'] > DatoTidStart) & (df1['Time'] < DatoTidSlutt)]  
    df2=df2[(df2['Time'] > DatoTidStart) & (df2['Time'] < DatoTidSlutt)]  
elif files == 3:
    df1=df1[(df1['Time'] > DatoTidStart) & (df1['Time'] < DatoTidSlutt)]  
    df2=df2[(df2['Time'] > DatoTidStart) & (df2['Time'] < DatoTidSlutt)] 
    df3=df3[(df3['Time'] > DatoTidStart) & (df3['Time'] < DatoTidSlutt)] 
#fjerner målinger gjort på andre tidspunkt 
#enn det vi ønsker basert på forhåndsbestemt start og slutt tid 
```

```python
format = '%I:%M%p'

if files == 1:
    df1['Time'] = pd.to_datetime(df1['Time'],format).apply(lambda x: x.time())
elif files == 2:
    df1['Time'] = pd.to_datetime(df1['Time'],format).apply(lambda x: x.time())
    df2['Time'] = pd.to_datetime(df2['Time'],format).apply(lambda x: x.time()) 
elif files == 3:
    df1['Time'] = pd.to_datetime(df1['Time'],format).apply(lambda x: x.time())
    df2['Time'] = pd.to_datetime(df2['Time'],format).apply(lambda x: x.time())
    df3['Time'] = pd.to_datetime(df3['Time'],format).apply(lambda x: x.time())

#dataset['TimeStamp'] = pd.Series([val.time() for val in dataset['TimeStamp']])
```

```python
if files == 1:
    df1=df1.set_index('Time') 
elif files == 2:
    df1=df1.set_index('Time')
    df2=df2.set_index('Time') 
elif files == 3:
    df1=df1.set_index('Time')
    df2=df2.set_index('Time')
    df3=df3.set_index('Time')
# Setter Tid som en index for radene
```

```python
if files == 1:
    print(df1)
elif files == 2:
    print(df1)
    print(df2)
elif files == 3:
    print(df1)
    print(df2)
    print(df3)
```
